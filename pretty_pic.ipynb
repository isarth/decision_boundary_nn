```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import pylab as pl
import random
```

```python
mean_1=[0,0]
cov_1=[[0.7,0.4],[0.4,0.7]]
dist_1=np.random.multivariate_normal(mean_1,cov_1,100)
mean_2=[2,2]
cov_2=[[1,0.2],[0.2,1]]
dist_2=np.random.multivariate_normal(mean_2,cov_2,100)
plt.scatter(dist_1[:,0],dist_1[:,1],color='green')
plt.scatter(dist_2[:,0],dist_2[:,1],color='red')
plt.show()
```

```python
data=np.zeros((200,2))
data[:100,:]=dist_1
data[100:,:]=dist_2
labels=np.zeros(200)
labels[100:]=1
```

```python
from sklearn.naive_bayes import GaussianNB
clf=GaussianNB()
clf.fit(data,labels)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf2=RandomForestClassifier()
clf2.fit(data,labels)
```

```python
from sklearn import svm
clf3=svm.SVC()
clf3.fit(data,labels)
```

```python
from sklearn import tree
clf4=tree.DecisionTreeClassifier()
clf4.fit(data,labels)
```

```python
clf.predict(data)
```

```python
def prettyPicture(clf, X_test, y_test):
    x_min = X_test[:,0].min(); x_max = X_test[:,0].max()
    y_min = X_test[ :,1].min(); y_max = X_test[:,1].max()

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, m_max]x[y_min, y_max].
    h = .01  # step size in the mesh
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    print xx[0][0]
    #print xx[1][0]
    print xx.shape,yy.shape
    xx2=xx
    yy2=yy
    
    #-----------------
    
    
    xx2=np.reshape(xx2,(xx.shape[0]*xx.shape[1]))
    yy2=np.reshape(yy2,(xx.shape[0]*xx.shape[1]))
    tester=np.zeros((xx.shape[0]*xx.shape[1],2))
    print xx2.shape
    tester[:,0]=xx2
    tester[:,1]=yy2
    Z=clf.predict(tester)
    #Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    
    # Put the result into a color plot
    print xx.shape
    Z = Z.reshape(xx.shape)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    #plt.pcolormesh(xx,yy,Z)
    plt.pcolormesh(xx, yy, Z, cmap=pl.cm.seismic)
    #plt.legend()
    #plt.show()
    # Plot also the test points
    grade_sig = [X_test[ii][0] for ii in range(0, len(X_test)) if y_test[ii]==0]
    bumpy_sig = [X_test[ii][1] for ii in range(0, len(X_test)) if y_test[ii]==0]
    grade_bkg = [X_test[ii][0] for ii in range(0, len(X_test)) if y_test[ii]==1]
    bumpy_bkg = [X_test[ii][1] for ii in range(0, len(X_test)) if y_test[ii]==1]

    plt.scatter(grade_sig, bumpy_sig, color = "b" )
    plt.scatter(grade_bkg, bumpy_bkg, color = "r")
    plt.legend()
    #plt.xlabel("bumpiness")
    #plt.ylabel("grade")
    plt.show()
```

```python
prettyPicture(clf4,data,labels)
```

```python
prettyPicture(clf3,data,labels)
```

```python
prettyPicture(clf2,data,labels)
```
